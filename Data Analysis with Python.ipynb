## is Total charges of services are biased?

```python
freq,bins=np.histogram(tel['TotalCharges'],bins=20)
```

```python
freq
```

```python
bins
```

```python
plt.figure(figsize=(10,6))
plt.hist(tel['TotalCharges'],bins=21,ec='k')
plt.xticks(bins,rotation=45)
plt.legend()
plt.show()
```

```python
## we know that Totacharges relationship between Tenure
## now we will analysing the correlation of both
#tenure Meaning - how many months a person has been a client of the company
```

```python
sns.heatmap(tel.corr(),annot=True)
plt.show()
```

```python
tel.corr()
```

```python
## Total charges and tenure have positive relation near of 1 and excat correlation point is 0.8 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(tel['tenure'],tel['TotalCharges'], color='blue')
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.title('Relationship Between No. of months a person has been client of the company and Total Charges',loc='center')
plt.show()
```

```python
# above we are observing,no. of months(tenure) increasing as well as TotalCharges also increasing.so,total charges is not biased
```
